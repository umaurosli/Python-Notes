```python
dff = dff.drop_duplicates()
dff = dff.drop(columns="Not_Useful_Column")
```

```python
dff
```

```python
dff["Last_Name"] = dff["Last_Name"].str.lstrip("...")
dff["Last_Name"] = dff["Last_Name"].str.lstrip("/")
dff["Last_Name"] = dff["Last_Name"].str.rstrip("_")
dff
```

```python
dff["Phone_Number"] = dff["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
dff["Phone_Number"] = dff["Phone_Number"].apply(lambda x : str(x))
dff["Phone_Number"] = dff["Phone_Number"].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:10])
dff["Phone_Number"] = dff["Phone_Number"].str.replace('nan--','')
dff["Phone_Number"] = dff["Phone_Number"].str.replace('Na--','')

dff
```

```python
dff[['Street_Address','State','Zipcode']] = dff['Address'].str.split(',',expand=True)
dff.drop(columns='Address')
```

```python
dff['Paying Customer'] = dff['Paying Customer'].str.replace('Yes','Y')
dff['Paying Customer'] = dff['Paying Customer'].str.replace('No','N')

dff['Do_Not_Contact'] = dff['Do_Not_Contact'].str.replace('Yes','Y')
dff['Do_Not_Contact'] = dff['Do_Not_Contact'].str.replace('No','N')
dff
```

```python
dff = dff.replace('N/a','')

dff = dff.fillna('')
dff
```

```python
for x in dff.index:
    if dff.loc[x,'Do_Not_Contact'] == 'Y' or dff.loc[x,'Do_Not_Contact'] == '':
        dff.drop(x, inplace=True)

dff
```

```python
dff = dff.reset_index(drop=True)
dff
```
